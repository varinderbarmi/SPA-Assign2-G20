```python
import json
from datetime import datetime
from kafka import KafkaConsumer
from datetime import datetime, timedelta
from kafka import KafkaProducer

    
```

```python

# Kafka consumer configuration for topic stock-market
consumer = KafkaConsumer("stock-market", bootstrap_servers=["localhost:9092"], value_deserializer=lambda x: json.loads(x.decode("utf-8")))

```

```python

# Kafka producer configuration for topic simple-moving-average
producer = KafkaProducer(bootstrap_servers=["localhost:9093"], value_serializer=lambda x: json.dumps(x).encode("utf-8"))
```

```python

# Send completed traded price after for all the intruments
def send_traded_price(instrument, cost, time):
    traded_price = {
        "instrument": instrument,
        "cost": cost,
        "time": time.isoformat()
    }
    producer.send("simple-moving-average", value=traded_price)
    print(f"Message sent to Kafka topic -> simple-moving-average sent :\n{traded_price}\n\n")

```

```python

# Dictionary to store current orders
orders = {}

```

```python


def clean_dict():
    key_to_remove = []
    for key, value in orders.items():
        # print(datetime.strptime(value["valid_until"], '%Y-%m-%dT%H:%M:%S.%f'))
        # print(datetime.now())
        if datetime.strptime(value["valid_until"], '%Y-%m-%dT%H:%M:%S.%f') <= datetime.now():
            key_to_remove.append(key)
    for i in key_to_remove:
        orders.pop(i)
        print("\nOrder No.: ", i," has expired. Hence, it is removed.\n")
            
            
```

```python
     
            

# Function to match orders
def match_orders():
    clean_dict()
    for order in list(orders.values()):
        #print(orders)
        #print(order["id"])
        if order["action"] == "BUY":
            match = [o for o in orders.values() if o["action"] == "SELL" and o["price"] == order["price"] and o["instrument"] == order["instrument"] ]
                    # and o["valid_until"] >= order["valid_until"] 
        elif order["action"] == "SELL":
            match = [o for o in orders.values() if o["action"] == "BUY" and o["price"] == order["price"] and o["instrument"] == order["instrument"] ]
                    # and o["valid_until"] >= order["valid_until"] 
        else:
            pass
        for m in match:
            if (m["quantity"] > order["quantity"] and m["id"] != order["id"] and m["instrument"] == order["instrument"] ):
                m["quantity"] -= order["quantity"]
                print(f"\n--------------------------\nTrade completed: \n--------------------------\n{order}\n <-> \n{m}\n--------------------------")
                print(f"Updated the Order No. ", m["id"], ". Changed the quantity to ", m["quantity"] )
                send_traded_price(m["instrument"] , m["price"], datetime.now())
                del orders[order["id"]]
            elif (m["quantity"] < order["quantity"] and m["id"] != order["id"] and m["instrument"] == order["instrument"] ):
                order["quantity"] -= m["quantity"]
                print(f"\n--------------------------\nTrade completed: \n--------------------------\n{order}\n <-> \n{m}\n--------------------------")
                print(f"Updated the Order No. ", order["id"], ". Changed the quantity to ", order["quantity"] )
                send_traded_price( order["instrument"] , order["price"], datetime.now())
                del orders[m["id"]]
            elif (m["quantity"] == order["quantity"] and m["id"] != order["id"] and m["instrument"] == order["instrument"] ):
                print(f"\n--------------------------\nTrade completed: \n--------------------------\n{order}\n <-> \n{m}\n--------------------------")
                send_traded_price(m["instrument"] , m["price"], datetime.now())
                del orders[m["id"]]
                del orders[order["id"]]
            else:
                pass
                
```

```python
     

# Kafka consumer loop
for count, msg in enumerate(consumer):
    try:
        order = msg.value
        #print(order)
        order["id"] = msg.offset
        #print(msg.offset)
        orders[order["id"]] = order
        #print(order)
        #print("Orders dict: \n",orders)
        #print("Current msg: \n",msg)
        match_orders()
    except:
        print("An exception occured with order: \n", order["id"], " --> ", order)

    
```
